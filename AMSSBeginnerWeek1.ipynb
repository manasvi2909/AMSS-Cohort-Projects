```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

path = "/content/city_day.csv"
```

```python
#reading dataset to dataframe and displaying number of rows and columns, and the first few entries
aqi_data = pd.read_csv(path)
print(aqi_data.shape)
print(aqi_data.head())
```

```python
#displaying a summary of the dataset and the datatypes in it
print(aqi_data.describe())
print(aqi_data.info())
```

```python
#printing the number of null values and their percentage
print(aqi_data.isnull().sum())
print(aqi_data.isnull().sum() / len(aqi_data) * 100) # Percentage missing
```

```python
#imputing NaN values in numerical columns
numerical_columns = ['PM2.5','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene','Xylene','AQI']
means = aqi_data[numerical_columns].mean()
aqi_data[numerical_columns] = aqi_data[numerical_columns].fillna(means)
print("Missing values after imputation:\n", aqi_data.isnull().sum())
```

```python
print(aqi_data['City'].value_counts())
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='City', data=aqi_data, order=aqi_data['City'].value_counts().index, palette='viridis')
plt.title('Number of Data Points per City')
plt.xlabel('Count')
plt.ylabel('City')
plt.show()
```

```python
print(aqi_data['AQI_Bucket'].value_counts())

plt.figure(figsize=(8, 5))
sns.countplot(x='AQI_Bucket', data=aqi_data, order=['Good', 'Satisfactory', 'Moderate', 'Poor', 'Very Poor', 'Severe'], palette='coolwarm')
plt.title('Distribution of AQI Buckets')
plt.xlabel('AQI Bucket')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

Delhi or other major metropolitan areas have the most data points. The AQI_Bucket distribution shows a prevalence of "Moderate" or "Poor" days in highly populated cities

```python
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_columns):
    plt.subplot(3, 5, i + 1)
    sns.histplot(aqi_data[col].dropna(), kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.tight_layout()
plt.show()
```

Benzene and CO are the least prevalent pollutants as can be seen by the lack of datatpoints and only data being for very low concentrations

```python
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_columns):
    plt.subplot(3, 5, i + 1)
    sns.boxplot(y=aqi_data[col].dropna())
    plt.title(f'Box Plot of {col}')
    plt.tight_layout()
plt.show()
```

Pollutants like SO2 show high density throughout concentration levels. It can also be inferred that Xylene has low density or lesser data points at higher concentrations.

```python
#relation between PM2.5 and AQI
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PM2.5', y='AQI', data=aqi_data, alpha=0.6)
plt.title('PM2.5 vs AQI')
plt.xlabel('PM2.5 Concentration')
plt.ylabel('AQI')
plt.grid(True)
plt.show()
```

There's a direct and strong positive relationship between increasing pollutant concentrations (especially PM2.5 and PM10) and rising AQI values
